```python
filenameParticles = '/Users/harabasz/Downloads/Thermal-FIST-master/input/list/PDG2020/list-withexcitednuclei.dat'
with open(filenameParticles,'r') as fileParticles:
    stringIO = io.StringIO(fileParticles.read().replace('#',''))
particles = pd.read_csv(stringIO,
                   skiprows=1,
                   delimiter=' ',
                   skipinitialspace=True)
```

```python
filenameDecays = '/Users/harabasz/Downloads/Thermal-FIST-master/input/list/PDG2020/decays.dat'
with open(filenameDecays,'r') as fileDecays:
    txtDecays = fileDecays.readlines()

lastLineEmpty = False
lineGroups = []
for line in txtDecays:
    line = line.replace('\n','')
    if len(line) == 0:
        lastLineEmpty = True
        continue
    if (line[0] == '#'):
        continue
    if lastLineEmpty:
        lineGroups.append([line.split()])
        lastLineEmpty = False
        continue
    lineGroups[-1].append(line.split())

lineGroups[:2]
```

```python
for group in lineGroups:
    if not int(group[1][0]) + 2 is len(group):
        print("Error: number of decay channels doesn't match the length of the group for {}",format(group[0][2]))
```

```python
parents = {}
for group in lineGroups:
    parent = (group[0][0], group[0][2])
    nchannels = int(group[1][0])
    for i in range(0,nchannels):
       # print(f"{i+1} of {nchannels}")
        channel = group[i+2]
        br = channel[0]
        children = []
        k = 1
       # print(channel)
        while '#' not in channel[k]:
            children.append(channel[k])
            k = k + 1
       # print(children)
        for child in children:
            if child not in parents:
                parents[child] = [parent,br,children]
            else:
                parents[child].append((parent,br,children))
print(len(parents))
parents
```
